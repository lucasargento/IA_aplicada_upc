```python
df = pd.get_dummies(df, columns=['Race','GenHealth','AgeCategory','Sex'])
df['HeartDisease'] = pd.Series(np.where(df.HeartDisease.values == 'Yes',1,0),df.index)
df['Smoking'] = pd.Series(np.where(df.Smoking.values == 'Yes',1,0),df.index)
df['AlcoholDrinking'] = pd.Series(np.where(df.AlcoholDrinking.values == 'Yes',1,0),df.index)
df['Stroke'] = pd.Series(np.where(df.Stroke.values == 'Yes',1,0),df.index)
df['DiffWalking'] = pd.Series(np.where(df.DiffWalking.values == 'Yes',1,0),df.index)
df['Diabetic'] = pd.Series(np.where(df.Diabetic.values == 'Yes',1,0),df.index)
df['PhysicalActivity'] = pd.Series(np.where(df.PhysicalActivity.values == 'Yes',1,0),df.index)
df['Asthma'] = pd.Series(np.where(df.Asthma.values == 'Yes',1,0),df.index)
df['KidneyDisease'] = pd.Series(np.where(df.KidneyDisease.values == 'Yes',1,0),df.index)
df['SkinCancer'] = pd.Series(np.where(df.SkinCancer.values == 'Yes',1,0),df.index)
```

```python
df.head()
```

```python
df.info()
```

Escalar:

```python
scaler = MinMaxScaler()
scaler.fit(df)
df_array = scaler.transform(df)
df_sc = pd.DataFrame(df_array,columns=df.columns,index=df.index)
```

**k-means with fixed k:** 

Inicializamos el algoritmo con k=4

```python
km_model = KMeans(n_clusters=4,random_state=3)
km_model.fit(df_sc)
```

Cluster centroids:

```python
centers = pd.DataFrame(km_model.cluster_centers_,columns=df_sc.columns)
```

```python
centers
```

```python
centers.std().sort_values()
```

Cluster labels:

```python
df_cluster = df_sc.copy()
df_cluster['cluster'] = km_model.labels_
```

```python
df_cluster
```

Cantidad de ejemplos en el cluster

```python
df_cluster['cluster'].value_counts()
```

```python
sse = []
for i in range(2, 11): 
       km = KMeans(n_clusters=i,random_state=0) 
       km.fit(df_sc) 
       sse.append(km.inertia_)
```


**Best number of clusters (Elbow Method):**



```python
visualizer = KElbowVisualizer(km, k=(2,12),timings=False)
visualizer.fit(df_cluster)
visualizer.show()
```

**Best number of clusters (calinski_harabasz):**

```python
visualizer = KElbowVisualizer(
    km, k=(2,12), metric='calinski_harabasz', timings=False
)

visualizer.fit(df_cluster) 
visualizer.show()  
```

**Best number of clusters (silhouette):**

```python
visualizer = KElbowVisualizer(
    km, k=(2,12), metric='silhouette', timings=False
)

visualizer.fit(df_cluster) 
visualizer.show() 
```
